```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data={"studyhour":[2,4,6,8,10],"Attendance":[70,80,85,90,95],"Assignment":[12,15,18,20,22],"FinalMark":[40,60,75,90,98]}
df=pd.DataFrame(data)
x=df[["studyhour","Attendance","Assignment"]]
y=df["FinalMark"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
# Train the Model
model = LinearRegression()
model.fit(x_train, y_train)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred = model.predict(x_test)
```

```python
# Evaluation Metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("\nEvaluation Metrics")
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python

new_data = [[6, 85, 18]]   # StudyHour=7, Attendance=88, Assignment=19
prediction = model.predict(new_data)

print("\nPrediction")
print(f"Predicted Final Marks: {prediction[0]:.2f}")
```

```python
import matplotlib.pyplot as plt

y_all_pred = model.predict(x)



plt.scatter(y, y_all_pred, color='black', label='Model Predictions')
plt.plot([y.min(), y.max()],
         [y.min(), y.max()],
         '--', color='black', linewidth=2)
plt.title("Actual vs Predicted Final Marks")
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.legend()
plt.grid(True)
```
